```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('dataset/social_network_ads.csv')
data.head()
```

```python
data.shape
```

```python
x=data.iloc[:,[2,3]].values #input data
y=data.iloc[:,4].values #output data
```

```python
x
```

```python
y
```

```python
#how to remove warning in jupyter notebook
import warnings;
warnings.simplefilter('ignore')
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
train_x
```

```python
test_x
```

```python
#perform feature scalling because input data:- 
#difference between age & salary much more....
#so that solving this problem we can optimizing this perform feature scalling...
```

```python
#load feature scalling library
from sklearn.preprocessing import StandardScaler
```

```python
s_c= StandardScaler()
train_x=s_c.fit_transform(train_x)
test_x=s_c.fit_transform(test_x)
```

```python
test_x
```

```python
#load KNN classifier 
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5,metric='euclidean',p=2)
knn.fit(train_x,train_y)
```

```python
y_pred=knn.predict(test_x)
y_pred
```

```python
test_y
```

```python
# using confusion matrix we can find out in testing data how many wrong & correct
#first load library of confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
c_m= confusion_matrix(test_y,y_pred)
c_m
```

```python
# [64+29=93] is correct prediction
# &[4=3=7] is wrong prediction then accuracy is 93% .
```

```python
import seaborn as sns
```

```python
sns.heatmap(c_m,annot=True,cbar=True)
plt.show()
```

```python
from sklearn.metrics import precision_score,recall_score
```

```python
precision_score(test_y,y_pred)
```

```python
recall_score(test_y,y_pred)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(test_y,y_pred)
```

# evaluation:-

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_y,y_pred)
```

```python
from matplotlib.colors import ListedColormap
x_set,y_set=test_x,test_y
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-1,
                            stop=x_set[:,0].max()+1,step=0.01),
                 np.arange(start=x_set[:,1].min()-1,
                          stop=x_set[:,1].max()+1,step=0.01))
plt.contour(x1,x2,knn.predict(np.array([x1.ravel(),
                                        x2.ravel()]).T).reshape(x1.shape),
            alpha=0.75,cmap=ListedColormap(('red','green')))
plt.xlim(x1.min(),x1.max())
plt.xlim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],
               c= ListedColormap(('red','green'))(i),label=j)
    
plt.title('knn testing dataset')
plt.xlabel('age')
plt.ylabel('estemated salary')
plt.legend()
plt.show()
```
